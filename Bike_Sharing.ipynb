<span style="color:black; font-size:30px;"> About the Company (Cyclistic)</span>

Cyclistic is a bike-share company launched in 2016 in Chicago. It has since grown to a fleet of 5,824 bicycles that are geo-tracked and locked into a network of 692 stations across Chicago. The bikes can be unlocked from one station and returned to any other station in the system anytime. Cyclistic offers a variety of pricing plans including: single-ride passes, full-day passes, and annual memberships. 

* **Customers who purchase a single-ride or full-day passes are referred to as casual riders.**
* **Customers who purchase annual memberships are Cyclistic members.**

Cyclisticâ€™s finance analysts have concluded that annual members are much more profitable than casual riders. Although the pricing flexibility helps Cyclistic attract more customers, Moreno (director of marketing) believes that maximizing the number of annual members will be key to future growth. Rather than creating a marketing campaign that targets all-new customers, Moreno believes there is a very good chance to convert casual riders into members. She notes that casual riders are already aware of the Cyclistic program and have chosen Cyclistic for their mobility needs. Now, the goal is to design marketing strategies aimed at converting casual riders into annual members. 

<span style="color:black; font-size:30px;"> Objective </span>

<span style="color:black; font-size:20px;"> Q: How do annual members and casual members use Cyclistic bikes differently?</span>
- The aim is to explore and understand how annual members and casual riders differ and why casual riders would buy a membership!

<span style="color:black; font-size:30px;"> Data </span>
##### **Bike trip data from June 2020 to May 2021 has been studied to understand the use of bikes by different users (member and casual).**  
Link to the data: https://divvy-tripdata.s3.amazonaws.com/index.html

The following information (columns) can be found in the data.

1. ride_id: Unique id assigned to a ride trip.
2. rideable_type: Type of bike used during trip by the user.
3. started_at: Start date and time for the trip
4. ended_at: End data and time for the trip
5. start_station_name: Name of the station where the trip started
6. start_station_id: Unique ID assigned to the start station.
7. end_station_name: Name of the station where the trip ended.
8. end_station_id: Unique ID assigned to the end station.
9. start_lat: Latitude of where the trip started.
10. start_lng: Longitude of where the trip started.
11. end_lat: Latitude of where the trip ended.
12. end_lng: Longitude of where the trip ended.
13. member_casual: User Type (either member or casual)

### Converting the start_name and end_time into datetime format. 

```python

file["start_time"]=pd.to_datetime(file["started_at"])
file["end_time"]=pd.to_datetime(file["ended_at"])
```

### Extract year, month, day, and day_name from 'start_time'

```python
# from the start_time, creating new columns: year, month, day and day_name for easier analysis.
file['year'] = file['start_time'].dt.year
file['month'] = file['start_time'].dt.month
file['day'] = file['start_time'].dt.day
file['day_name'] = file['start_time'].dt.day_name()
file['hour'] = file['start_time'].dt.hour
```

### Mapping the day names to numerical values (1-7)

### Determining the number of member and casual riders.

```python

casual_riders = file[file['member_casual'] == 'casual']['ride_id'].count()
member_riders = file[file['member_casual'] == 'member']['ride_id'].count()

print("Number of Casual Riders:", casual_riders)
print("Number of Member Riders:", member_riders)
```

### Determining the mean **ride_length** of (1: all user, 2: member user, and 3: casual user)

```python
mean_rl_total=file["ride_length"].mean()
mean_ride_length = file.groupby('member_casual')['ride_length'].mean()

print('The mean ride_lenght is:',mean_rl_total)
print()
print('The mean ride_lenght for each user type is:', mean_ride_length)

# Creating a horizontal bar chart
plt.figure(figsize=(8, 3))

# Plotting a horizontal bar chart
bars=plt.barh(mean_ride_length.index, mean_ride_length, color=['gray', 'orange'])
plt.title('Mean Ride Length by User Type')
plt.xlabel('Mean Ride Length')
plt.ylabel('User Type')
plt.tight_layout()
plt.show()
```

### Determining the max ride_lenght of rides from (1: all user, 2: member user, and 3: casual user)

```python
max_rl_total=file["ride_length"].max()
max_ride_length = file.groupby('member_casual')['ride_length'].max()

print('The max ride_lenght (for all users) is:', max_rl_total)
print()
print('The max ride_lenght for each user type is:', max_ride_length)

# Creating a horizontal bar chart
plt.figure(figsize=(8, 3))

# Plotting a horizontal bar chart
bars=plt.barh(max_ride_length.index, max_ride_length, color=['gray', 'orange'])
plt.title('Max Ride Length by User Type')
plt.xlabel('Max Ride Length')
plt.ylabel('User Type')
plt.tight_layout()
plt.show()
```

### Determining the mode of day of week for (1: all user, 2: member user, and 3: casual user)

```python
# Calculating the mode of week_day for all types of users
mode_week_day = file["day_name"].mode()
print(mode_week_day)

# Calculating the mode of week_day for member user type
mode_week_day_member = file[file["member_casual"]=="member"]
mode_week_day_member = mode_week_day_member["day_name"].mode()
print(mode_week_day_member )

# Calculating the mode of week_day for casual user type
mode_week_day_casual = file[file["member_casual"]=="casual"]
mode_week_day_casual = mode_week_day_casual["day_name"].mode()
print(mode_week_day_casual)
```

### Determining the number of member and casual riders each day of the week. 

```python
rides_user_type = file.groupby(['day_name', 'member_casual'])['ride_id'].count().reset_index()

pivot_table = rides_user_type.pivot(index='day_name', columns='member_casual', values='ride_id')

print(pivot_table)
```

### Bar-graph showing the number of riders (members and casuals) for each day of the week.

```python
pivot_table = pivot_table.loc[['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']] ## loc is used to reorder the index of days to start from Sunday.

pivot_table.plot(kind='bar', figsize=(12, 5), color=['gray', 'orange'])
plt.title('Number of Riders for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Riders')
plt.legend(title='User Type')
plt.xticks(rotation=0) 
plt.tight_layout()
plt.show()
```

### Determining the average ride_lenght of member and casual riders for each day of the week

```python
avg_ride_per_day = file.groupby(['day_name', 'member_casual'])['ride_length'].mean().unstack()

ordered_days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
ordered_data = avg_ride_per_day.loc[ordered_days]

print(ordered_data)
```

### Bar-graph showing the average ride lenght for each day of the week for riders (members and casuals).

```python
avg_ride_per_day = avg_ride_per_day.loc[['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']]

# Plotting the bar graph
avg_ride_per_day.plot(kind='bar', figsize=(12, 5), color=['gray', 'orange'])
plt.title('Average Ride Length for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Ride Length')
plt.legend(title='User Type')
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

### Bar-graph showing the number of rides per month for member and casual riders.

```python
file['year_month'] = file['start_time'].dt.to_period('M')  # extracting the Year-Month in 'Period' format
rides_per_month_user = file.groupby(['year_month', 'member_casual'])['ride_id'].count().reset_index()

pivot_table = rides_per_month_user.pivot(index='year_month', columns='member_casual', values='ride_id')

pivot_table.sort_index(inplace=True) # Sorting the 'year_month' index in chronological order

pivot_table.plot(kind='bar', figsize=(14, 6), color=['gray', 'orange'])
plt.title('Number of Rides per Month for Member and Casual Riders')
plt.xlabel('Year, Month')
plt.ylabel('Number of Rides')
plt.legend(title='User Type')
plt.xticks(rotation=0) 
plt.tight_layout()
plt.show()
```

### Line Graph showing hte number of rides per hour for casual and member users.

```python

rides_per_hour_user = file.groupby(['hour', 'member_casual'])['ride_id'].count().reset_index()

pivot_table = rides_per_hour_user.pivot(index='hour', columns='member_casual', values='ride_id').fillna(0)
plt.figure(figsize=(14, 6))
plt.plot(pivot_table.index, pivot_table['casual'], marker='o', label='Casual', color='skyblue')
plt.plot(pivot_table.index, pivot_table['member'], marker='o', label='Member', color='orange')

plt.title('Number of Rides per Hour for Casual and Member Users')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Line Graphs for rides per hour per day for casual and rider user (7 different plots)

```python
rides_per_hour_day = file.groupby(['hour', 'day_name', 'member_casual'])['ride_id'].count().reset_index() # grouping by hour, day_name and member_casual

fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(16, 8)) # as there are 7 days, creating subplots arranged in two rows. 
days_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for i, day in enumerate(days_of_week): # loop over each day (current day)
    day_data = rides_per_hour_day[rides_per_hour_day['day_name'] == day]
    
    pivot_table = day_data.pivot_table(index='hour', columns='member_casual', values='ride_id').fillna(0) # grouping for the current day.
    
    row = i // 4  
    col = i % 4   
    
    ax = axs[row, col]
    ax.plot(pivot_table.index, pivot_table['casual'], marker='o', label='Casual', color='gray')
    ax.plot(pivot_table.index, pivot_table['member'], marker='o', label='Member', color='orange')
    ax.set_title(f'{day}')
    ax.set_xlabel('Hour of the Day')
    ax.set_ylabel('Number of Rides')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
